```python
x=newData['Amount']
```

```python
labels=newData.index
```

```python
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(pct*total/100.0)
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
```

```python
figure(figsize=(8, 8), dpi=80)
plt.pie(x, labels=labels,autopct=make_autopct(x))
plt.show()
```

```python
def Monthly_Report(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df[(df['Date'] >='2022-02-10') & (df['Date'] <= '2022-02-22')]
    df = df[(df['Amount'] > 0)]
    print("This month you have spend {}".format(df['Amount'].sum()))
    # df.groupby(['Category','Description']).sum()
    newData = df.groupby(['Category']).sum()
    print(newData)
    df.groupby(['Category','Description']).sum()
    x=newData['Amount']
    labels=newData.index
    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(pct*total/100.0)
            return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
        return my_autopct
    figure(figsize=(8, 8), dpi=80)
    plt.pie(x, labels=labels,autopct=make_autopct(x))
    plt.show()
```

```python
Monthly_Report(df)
```
