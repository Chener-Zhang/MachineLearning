```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def Monthly_Report(df,begin,end):
    df['Date'] = pd.to_datetime(df['Date'])
    df = df[(df['Date'] >=begin) & (df['Date'] <= end)]
    df = df[(df['Amount'] > 0)]
    print("This month from {} to {} you have spend {}".format(begin,end,df['Amount'].sum()))
    # df.groupby(['Category','Description']).sum()
    newData = df.groupby(['Category']).sum()
    print(newData)
    df.groupby(['Category','Description']).sum()
    print(df.groupby(['Category','Description']).sum())
    x=newData['Amount']
    labels=newData.index
    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(pct*total/100.0)
            return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
        return my_autopct
    figure(figsize=(8, 8), dpi=80)
    plt.pie(x, labels=labels,autopct=make_autopct(x))
    plt.show()
    
```

```python
activity = pd.read_csv("activity.csv")
Monthly_Report(activity,'2022-02-22','2022-03-16')
```
