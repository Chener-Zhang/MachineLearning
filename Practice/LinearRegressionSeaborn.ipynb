```python
def costFunction(m,b,X,y):
    y_expected = y - m*X +b 
    residual_square = (y - y_expected)**2
    sum_residual_square = sum(residual_square)
    mean_square_error = sum_residual_square/len(X)
    return mean_square_error
```

```python
def costFunction_M(X,y):
    number = np.arange(0,80)/1000 
    fig = plt.figure()
    ax = fig.add_subplot(111, projection = '3d')
    for m in number:
        b = m * 1000
        ax.scatter(m, b, costFunction(m,b,X,y))
```
